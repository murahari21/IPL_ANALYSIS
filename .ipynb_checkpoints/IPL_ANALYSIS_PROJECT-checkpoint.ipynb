```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import sessionmaker

```

```python
matches_df = pd.read_csv("matches.csv")
deliveries_df = pd.read_csv("deliveries.csv")
players_df = pd.read_csv("players.csv")

print(matches_df.head())
print(deliveries_df.head())
print(players_df.head())
```

```python
matches_df.fillna("Unknown", inplace=True)
deliveries_df.fillna(0, inplace=True)
players_df.fillna("Unknown", inplace=True)
```

```python
matches_df['match_date'] = pd.to_datetime(matches_df['match_date'])
```

```python
deliveries_df['batsman_runs'] = deliveries_df['batsman_runs'].astype(int)
deliveries_df['bowler_runs'] = deliveries_df['bowler_runs'].astype(int)
deliveries_df['is_wicket'] = deliveries_df['is_wicket'].astype(int)
```

```python
player_runs = deliveries_df.groupby('match_id')['batsman_runs'].sum().reset_index()
player_runs.columns = ['match_id', 'total_match_runs']
```

```python
match_wickets = deliveries_df.groupby('match_id')['is_wicket'].sum().reset_index()
match_wickets.columns = ['match_id', 'total_wickets']
```

```python
match_summary = matches_df.merge(player_runs, on="match_id")
match_summary = match_summary.merge(match_wickets, on="match_id")

print(match_summary.head())
```

```python
engine = create_engine("sqlite:///ipl_etl.db")
Base = declarative_base()
```

```python
matches_df.to_sql("matches", engine, if_exists="replace", index=False)
deliveries_df.to_sql("deliveries", engine, if_exists="replace", index=False)
players_df.to_sql("players", engine, if_exists="replace", index=False)
match_summary.to_sql("match_summary", engine, if_exists="replace", index=False)
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
from sqlalchemy import text
```

```python
result = session.execute(text("""
SELECT season, COUNT(*) as total_matches
FROM matches
GROUP BY season
ORDER BY season;
""")).fetchall()

print(result)
```

```python
wins = session.execute(text("""
SELECT winner, COUNT(*) as total_wins
FROM matches
GROUP BY winner
ORDER BY total_wins DESC;
""")).fetchall()

print(wins)
```

```python
high_score = session.execute(text("""
SELECT match_id, total_match_runs
FROM match_summary
ORDER BY total_match_runs DESC
LIMIT 1;
""")).fetchall()

print(high_score)
```

```python
total_wickets = session.execute(text("""
SELECT SUM(is_wicket) FROM deliveries;
""")).fetchall()

print(total_wickets)
```

```python
pd.read_sql("""
SELECT match_id, SUM(batsman_runs) AS total_runs
FROM deliveries
GROUP BY match_id
ORDER BY total_runs DESC
LIMIT 5
""", engine)
```

```python
pd.read_sql("""
SELECT m.season, AVG(d.batsman_runs) AS avg_runs
FROM deliveries d
JOIN matches m ON d.match_id = m.match_id
GROUP BY m.season
ORDER BY m.season
""", engine)
```

```python
pd.read_sql("""
SELECT m.season, SUM(d.is_wicket) AS total_wickets
FROM deliveries d
JOIN matches m ON d.match_id = m.match_id
GROUP BY m.season
ORDER BY m.season
""", engine)
```

```python
import matplotlib.pyplot as plt

season_runs = pd.read_sql("""
SELECT m.season, SUM(d.batsman_runs) AS total_runs
FROM deliveries d
JOIN matches m ON d.match_id = m.match_id
GROUP BY m.season
""", engine)

plt.figure(figsize=(8,5))
plt.plot(season_runs['season'], season_runs['total_runs'])
plt.title("Total Runs Per Season")
plt.xlabel("Season")
plt.ylabel("Total Runs")
plt.show()
```
